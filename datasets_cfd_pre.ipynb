```python
def getArr(start, stop):
    return list(range(start, stop))

r = [0,1,2,
     4,6,7,
     14,18,
     19,21,23,25,27,29,31,33,35,37,39,
     43,45,47,49,
     *getArr(57,75),76,77,78,
     *getArr(81,88),*getArr(103,117)]
df.iloc[:, r].head()
```

```python
import os
from glob import glob

a = []
b = []
for path in glob("../../datasets/1/face/0_*"):
    filename = os.path.basename(path)
    if not filename.endswith("-N.jpg"):
        continue
    a.append(filename)
    filename = filename.split("-N.jpg")[0].split("CFD-")[1]
    a_ = filename.split("-")

    m = "-".join(a_[0:2])
    if a_[0] == "IF" or a_[0] == "IM":
        m = "{}{}-{}".format(a_[0], a_[1], a_[2])
    b.append("{}{}".format(m, "-1" if len(a_) == 4 and len(a_[3]) == 1 else ""))
pd.DataFrame({"filename": a, "model_id": b}).merge(df.iloc[:, r], how="left", on="model_id").to_csv("../../datasets/1/label_used.csv", quoting=1, encoding="utf-8-sig", index=False)
```
